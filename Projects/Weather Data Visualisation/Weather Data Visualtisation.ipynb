## Data Preparation & Cleaning

While the records contain a wealth of information, we'll limit our analysis to the following areas:

Monthly, yearly distribution of rain

Temparature, Humidity and Visibility relationship

Wind Speed and Wind Bearing connections

Let's select a subset of columns with the relevant data for our analysis.

Looking at the dataset, the columns 'Precip Type' and 'Loud Cover' have the same value throughout the dataset. The value for Precip Type is rain and the value for Loud Cover througout the dataset is zero.

Moreover, The column 'Daily Summary' seems irrelevant as it is a repitition of the column 'Summary' which meks it not so useful. We'll be dropping this col as well.

No concrete conclusion can be found by using these 3 columns so it it'll be better to just drop them.

```python
 weather_df.drop(['Precip Type', 'Loud Cover', 'Daily Summary'], axis = 1,inplace = True)
```

axis = 1 is to specify that the column should be discarded and inplace = True signifies that no new dataset must be created and to make changes in the already existing one.

```python
weather_df
```

Now that all the unnecesary values are removed, we can move further and get to the data preparation part. 

The column,'Formatted Date', has date and time together. We'll be extracting them and storing them in separate columns for even efficient data visualisation.

In all the numeric field values, no precision is set, so we'll be setting the precision to 3 decimal places in all the columns' data.

```python
weather_df['Formatted Date'] = weather_df['Formatted Date'].str[:-6]
weather_df
```

Extracting the date and time from the column

```python

weather_df['Formatted Date'] = pd.to_datetime(weather_df['Formatted Date'])
weather_df['Year'] = weather_df['Formatted Date'].dt.year
weather_df['Month'] = weather_df['Formatted Date'].dt.month
weather_df['Day'] = weather_df['Formatted Date'].dt.day
weather_df['Hour'] = weather_df['Formatted Date'].dt.hour
```

```python
weather_df
```

Now, let's round all the numerical data values to the nearest 3 decimal places

```python
weather_df[['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)']] = weather_df[['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)']].round(3)
```

```python
weather_df
```

Now that we've successfully extracted the day,month,year and the hours of the day, we can make the data visualisation much more specific and create better visualisation plots.

### Comparison Plots for temperature vs month

```python
import calendar
fig, axs = plt.subplots(4, 3, figsize=(15, 15))

for month in range(1, 13):
    month_data = weather_df[weather_df['Month'] == month]

    row = (month - 1) // 3
    col = (month - 1) % 3

    sns.lineplot(data=month_data, x='Day', y='Temperature (C)', ax=axs[row, col])
    axs[row, col].set_title(calendar.month_name[month])
    axs[row, col].set_xlabel('Day')
    axs[row, col].set_ylabel('Temperature (C)')

plt.tight_layout()
plt.show()
```

The above plot gives us he visualisation of the distribution of temperatures across the months of a year.

## Humidity vs Visibility

Humidity refers to the amount of moisture present in the air, typically expressed as a percentage. Visibility represents the clarity and extent to which objects can be seen in the atmosphere, often measured in kilometers (km).

Higher humidity levels can lead to the formation of moisture in the air, which can affect visibility. Higher humidity can contribute to fog or mist formation, reducing visibility.

Let's plot a graph and seee if the above statement is true

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=weather_df, x='Humidity', y='Visibility (km)', hue='Visibility (km)', palette='viridis')

sns.regplot(data=weather_df, x='Humidity', y='Visibility (km)', scatter=False, color='black')

plt.title('Humidity vs Visibility')
plt.xlabel('Humidity')
plt.ylabel('Visibility (km)')

plt.show()
```

The graph reveals the correlation between humidity and visibility. It shows how changes in humidity impact visibility levels.

As humidity increases, visibility tends to decrease. This indicates that higher levels of moisture in the air can contribute to reduced visibility.

The relationship between humidity and visibility may vary depending on other factors such as temperature, weather conditions, and the presence of fog, mist, or precipitation.
